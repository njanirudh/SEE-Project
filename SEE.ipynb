```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python

data = pd.read_excel('Results.xlsx',sheet_name=0,stylesheet="abc")
newdata = data.as_matrix()
x = newdata[2:,0]
print (x)
y = newdata[2:,1]
theta = newdata[2:,4]
pre=[]
for i in x:
    pre.append(i-701.85)
print (np.sum(pre)/20)
```

```python
#X_Right
print (x)
x_mean=np.mean(x)
x_var=np.var(x)
print (x_var)
y_mean=np.mean(y)
theta_mean=np.mean(theta)
plt.hist(x[np.newaxis].T,bins=5,density=True,color='y')
x.shape
x_std=np.sqrt(x_var)
x_right=np.linspace(x_mean-3*x_std,x_mean+3*x_std,20)
plt.plot(x_right,stats.norm.pdf(x_right,x_mean,x_std))
plt.title("X-Right")
plt.xlabel('X-Right in mm')
plt.ylabel('number of counts')
plt.grid()
plt.show()
#y
print (y)
y_mean=np.mean(y)
y_var=np.var(y)
print (y_mean)
print (y_var)
plt.hist(y[np.newaxis].T,bins=5,density=True,color='r')
y_std=np.sqrt(y_var)
y_left=np.linspace(y_mean-3*y_std,y_mean+3*y_std,20)
plt.plot(y_left,stats.norm.pdf(y_left,y_mean,y_std))
plt.title("Y-Right")
plt.xlabel('Y-Right in mm')
plt.ylabel('number of counts')
plt.grid()
#plt.xlim([450,600])
plt.show()
#theta
theta_mean=np.mean(theta)
theta_var=np.var(theta)
print(theta_var)
plt.hist(theta[np.newaxis].T,bins=5,density=True, color='g')
theta_std=np.sqrt(theta_var)
theta_left=np.linspace(theta_mean-3*theta_std,theta_mean+3*theta_std,20)
plt.plot(theta_left,stats.norm.pdf(theta_left,theta_mean,theta_std))
plt.grid()
plt.xlabel('angle in degrees')
plt.ylabel('number of counts')
plt.title("Angle")
plt.show()
```

**Accuracy** <br>
To calculate accuracy we need to understand the ground true value(true pose) and the actual value (achieved pose). Thus accuracy is the amount of deviation from ground truth to achieved result.In our experiment, accuracy is the deviation of the actual pose to the true pose.<br>
Calculation of actual pose is given by the following:<br>
* Given: <br>
SPEED=180.00mm/s<br> 
SLEEP_TIME= 3s <br>
TURN_RATE= 10% <br>
Thus the robot has three end poses accordingly straight, right and left end pose.<br>
The true straight pose is given by-<br>
$$d=SPEEDxTIME$$
therefore $$ d=180*3= 540mm $$
* Obtaining left or right true pose:<br>
Since the robot is a differential drive robot, varying the velocity of each wheel will make the robot to perform rotation about a point called the instantaneous center of rotation(ICC).<br>
Given velocity of each wheel ($V_r,V_l$) and distance between two wheels ($l$). 
$$V_r =180mm/s$$
$$V_l= \frac{180}{100}(180-10)=162mm/s$$
$$l=120mm$$
To have ($R$) which is the distance from the ICC to the mid-point between the wheels($l$)<br>
$$R=[\frac{l}{2}][\frac{V_l+V_r}{V_r-V_l}]$$
$$l=120mm$$
$$ R= [\frac{120}{2}][\frac{162+180}{180-162}] = 1140mm$$
For the rotation rate $ω$
$$ω=\frac{V_r-V_l}{l}$$
$$ω= \frac{180-162}{120}=0.15rad/s$$
thus for true right d = (ω)(SLEEP_TIME)= 0.15 x 3 = 0.45rad= 25.7831deg<br>
Using this ICC, we project the corresponding angle onto the grid sheet as show in the figure below. Since our marker is located in the center of the axle, the true pose is projected onto the sheet, using the following steps:<br>
* Place the robot onto to the designated origin position, as shown in the figure().
* Accordingly from the position of the marker, measure the distance R with respect to the below figure.
* Using basic geometric apparatus, draw a quadrant of a circle, such that it begins at the marker position and follows upto the rotation rate angle(d).
* Resulting is the true end pose for right/left trajectories.
